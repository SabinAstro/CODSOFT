data analysis with the separated data

0 for legal transactions
1 for fraud transaction

```python
legal = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```


Statistics for legal data

```python
legal.Amount.describe()
```

comparing both the values

```python
credit_data.groupby('Class').mean()
```

under sampling data

build similar dataset containing both transcations

no.of fraud transcations=492

```python
legal_sample=legal.sample(n=492)
```

Visulaization

```python
credit_data['Class'].value_counts()
```

```python
sns.countplot(x='Class',data=credit_data)
```

concatenate two dataset

```python
new_data=pd.concat([legal_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

splitting the targets and other features

```python
x=new_data.drop(columns='Class',axis=1)
y=new_data['Class']
```

```python
print(x)
```

```python
print(y)
```

splitting the training data and testing data separately

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

training the model

using the logistic regression

```python
model=LogisticRegression()
```

training the model with training data

```python
model.fit(x_train,y_train)
```

evaluating the model

Accuracy score

```python
x_train_predict=model.predict(x_train)
train_data_acc=accuracy_score(x_train_predict,y_train)
```

```python
print(x_train_predict)
```

```python
print("accuracy of training data:",train_data_acc)
```

testing the data with model

```python
x_test_predict=model.predict(x_test)
test_data_acc=accuracy_score(x_test_predict,y_test)
```

accuracy of test data

```python
print("accuracy of test data:",test_data_acc)
```

testing the model with random forest

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
RF=model.predict(x_test)
accuracy_score(RF,y_test)
```
